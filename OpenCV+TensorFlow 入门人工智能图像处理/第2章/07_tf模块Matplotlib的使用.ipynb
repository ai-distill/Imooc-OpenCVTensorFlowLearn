```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([3, 5, 7, 6, 2, 6, 10, 15])
```

```python
# 通过plot方法绘折线图，需要传入参数
# 第1个参数是x轴坐标
# 第2个参数是y轴坐标
# 第3个参数可以给颜色
plt.plot(x, y, 'r')
```

```python
# 三大图表
# 折线， 饼状，柱状
```

```python
# 最后一个参数指定线条的宽度
plt.plot(x, y, 'g', lw=10)
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([13, 25, 17, 36, 21, 16, 10, 15])
```

```python
# 绘制柱状图
# 第1个属性表明横坐标的值
# 第2个属性表明纵坐标
# 第3个属性表明每个柱状图占用宽度比例
# 第4个属性表明透明度 
# 第5个属性表明颜色
plt.bar(x, y, 0.5, alpha=1, color='b')
plt.show()
```
